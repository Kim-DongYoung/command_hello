```python
#   
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns # Pairplot

import copy



# Configuration
# columns_N :  
columns_N = 8



# Zero phase filtering Parameter
ORDER = 5
window = 12
#delay = 5
delay = 5

np_ZEP = window + 10 * ORDER

Transfer_Function_Value1 = np.array([1., -3.98454312, 6.43486709, -5.25361517, 2.16513291, -0.359928245])
Transfer_Function_Value2 = np.array([0.000059795780369994, 0.000298979, 0.000597958, 0.000597958, 0.000298979, 0.000059795780369994])



#train_x_orig, train_y, test_x_orig, test_y, classes = load_data()

df_Data = pd.DataFrame()
df_Data = pd.read_csv('./data/190805_RSSI/190131_RSSI data.csv', encoding='CP949')
df_Data.columns = ['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7', 
                  'Direction', 'Distance']

cols_Module = ['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7']
cols_row = ['Distance']

X_Data = df_Data[cols_Module] 
X_Ideal_Data = df_Data[cols_Module]
X_row_Data = df_Data[cols_row]

X_count_row = X_row_Data.shape[0]

df_X_Data = pd.DataFrame(np.array(X_Data), columns=['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7'])
df_X_Ideal_Data = pd.DataFrame(np.array(X_Ideal_Data), columns=['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7'])
df_X_row_Data = pd.DataFrame(np.array(X_row_Data), columns=['Distance'])

y_ZEP_1st = pd.DataFrame(np.zeros((np_ZEP, columns_N)), columns=['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7'])
y_ZEP_2nd = pd.DataFrame(np.zeros((np_ZEP, columns_N)), columns=['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7'])


print(X_Data.head())
print(X_Ideal_Data.tail())
print(df_X_Data.head())
print(df_X_Ideal_Data.tail())

# Show all data
plt.plot(X_Data)
plt.show()



# (Zero phase) 
# -      
def zerophase_filter(ORDER, Transfer_Function_Value1, Transfer_Function_Value2, np_ZEP, X, y_ZEP):
    """
    ORDER = 5   
    Transfer_Function_Value1
    Transfer_Function_Value2
    np_ZEP = window + 10 * ORDER  #   
    X = 1st filt : rawdata_window[j] -> 2nd filt : zero_phase_result[j]
    """    
 
    y_ZEP[0:1] = Transfer_Function_Value2[0:1] * X[0:1]
    for i in range (1, ORDER + 1):
        y_ZEP[i:i+1] = 0
        
        #for j in range (i):
        for j in range (i + 1):
            y_ZEP[i:i+1] = y_ZEP[i:i+1] + Transfer_Function_Value2[j:j+1] * X[i - j:i - j +1]
            
        #for j in range (i - 1):
        for j in range (i):
            y_ZEP[i:i+1] = y_ZEP[i:i+1] - Transfer_Function_Value1[j + 1:j + 2] * y_ZEP[i - j - 1:i - j]

    # ORDER + 1  np_ZEP .         
    for i in range (ORDER + 1, np_ZEP + 1):
        y_ZEP[i:i+1] = 0
        
        #for j in range (ORDER):
        for j in range (ORDER + 1):
            y_ZEP[i:i+1] = y_ZEP[i:i+1] + Transfer_Function_Value2[j:j+1] * X[i - j:i - j +1]
            
        #for j in range (ORDER - 1):
        for j in range (ORDER):
            y_ZEP[i:i+1] = y_ZEP[i:i+1] - Transfer_Function_Value1[j + 1:j + 2] * y_ZEP[i - j - 1:i - j]

            
def filtfilt(ORDER, Transfer_Function_Value1, Transfer_Function_Value2, np_ZEP, X, y_ZEP_1st, y_ZEP_2nd):
    
    zerophase_filter(ORDER, Transfer_Function_Value1, Transfer_Function_Value2, np_ZEP, X, y_ZEP_1st)
    
    y_ZEP_1st = np.array(y_ZEP_1st)
    y_ZEP_1st_out = np.array(y_ZEP_1st)
    
    for i in range (np_ZEP):
        y_ZEP_1st_out[i:i+1] = y_ZEP_1st[np_ZEP - i - 1:np_ZEP - i]
        
    #y_ZEP_1st_out = pd.DataFrame(np.array(y_ZEP_1st_out), columns=['ToF1_Main', 'ToF2_Sub5', 'ToF3_Sub6', 'ToF4_Sub7'])
    
    zerophase_filter(ORDER, Transfer_Function_Value1, Transfer_Function_Value2, np_ZEP, y_ZEP_1st_out, y_ZEP_2nd)
    
    y_ZEP_2nd = np.array(y_ZEP_2nd)
    y_ZEP_2nd_out = np.array(y_ZEP_2nd)
    
    for i in range (np_ZEP):
        y_ZEP_2nd_out[i:i+1] = y_ZEP_2nd[np_ZEP - i - 1:np_ZEP - i]    
        
    #y_ZEP_2nd_out = pd.DataFrame(np.array(y_ZEP_2nd_out), columns=['ToF1_Main', 'ToF2_Sub5', 'ToF3_Sub6', 'ToF4_Sub7'])
    
    #  : y_ZEP_2nd_out[window - 1 - delay]
    #     : y_ZEP_2nd_out[window - 1 - 1] #  :       .
    return y_ZEP_2nd_out[window - 2 - delay]
    #return y_ZEP_2nd_out



#    np.array   
#  (B )   

X = np.array(df_X_Data)
X_row_Data = np.array(X_row_Data)

X_Input = np.zeros((np_ZEP, columns_N))
y_ZEP_result_temp = np.array(X)
y_ZEP_result = np.array(X)

Transfer_Function_Value1 = np.array(Transfer_Function_Value1)
Transfer_Function_Value2 = np.array(Transfer_Function_Value2)

y_ZEP_1st = np.array(y_ZEP_1st)
y_ZEP_2nd = np.array(y_ZEP_2nd)

#zero_window = np.zeros((X_count_row,1))
df_zero_window = pd.DataFrame(np.zeros((1,1)), columns=['Distance'])
window_jump = 0
p = 0
# Correction_value = 2
# 
Correction_value = 100/96



# X_count_row    
#        
#      11  12~22  .

for i in range ((X_count_row+1) - window):
    
    #  (12)      
    for j in range (window):
        k = i + j
        X_Input[j:j + 1] = X[k:k + 1]            
    
    # X_count_row     
    #   (Correction_value) = -2
    #if (k-1)%100 > 10:
    if X_row_Data[k:k+1] == X_row_Data[k-1:k]:
        if (window_jump == 0):
            y_ZEP_result_temp[i + window - 1 :i + window] = filtfilt(ORDER, Transfer_Function_Value1, Transfer_Function_Value2, np_ZEP, X_Input, y_ZEP_1st, y_ZEP_2nd) 
            
            y_ZEP_result[i + window - 1 :i + window] = y_ZEP_result_temp[i + window - 1 :i + window] * Correction_value
            
            #   
            #if (y_ZEP_result_temp[i + window - 1 :i + window] > 0):
            #    y_ZEP_result[i + window - 1 :i + window] = y_ZEP_result_temp[i + window - 1 :i + window] + Correction_value
            
            #elif (y_ZEP_result_temp[i + window - 1 :i + window] < 0):
            #    y_ZEP_result[i + window - 1 :i + window] = y_ZEP_result_temp[i + window - 1 :i + window] - Correction_value
                
            #else:
            #    y_ZEP_result[i + window - 1 :i + window] = y_ZEP_result_temp[i + window - 1 :i + window]
                
        #    
        elif (window_jump > 0): 
            window_jump = window_jump - 1
            #y_ZEP_result[i - window: i - window + 1] = y_ZEP_result[i : i + 1] 
                    
    #       
    elif X_row_Data[k:k+1] != X_row_Data[k-1:k]:
        df_zero_window = df_zero_window.set_value(len(df_zero_window), 'Distance', k)
        
        window_jump = window - 1
        
        #    
        #   11  12~22  
        Start_row = np.int64(pd.to_numeric(df_zero_window['Distance']))
        y_ZEP_result[Start_row[p] : Start_row[p] + window] = y_ZEP_result[Start_row[p] + window : Start_row[p] + window + window]
        p = p + 1
        
#        11  12~22   
y_ZEP_result[Start_row[p] : Start_row[p] + window] = y_ZEP_result[Start_row[p] + window : Start_row[p] + window + window]



#Start_row = list(map(int, np.array(df_zero_window[1:2])))
#p = p + 1
#y_ZEP_result[Start_row : Start_row + 12] = y_ZEP_result[Start_row + 12 : Start_row + 24]   

#Start_row = np.int64(pd.to_numeric(df_zero_window['Distance']))
#y_ZEP_result[Start_row[0] : Start_row[0] + 12] = y_ZEP_result[Start_row[0] + 12 : Start_row[0] + 24]   

#Start_row = np.int64(df_zero_window)
#Start_row
#y_ZEP_result[Start_row[0] : Start_row[0] + 12] = y_ZEP_result[Start_row[0] + 12 : Start_row[0] + 24]   

Start_row = np.int64(pd.to_numeric(df_zero_window['Distance']))
Start_row



# pandas dataframe
# X_ZEP y_ZEP_result[window - 1 - delay]  
#   .
#y_ZEP_result = pd.DataFrame(np.array(y_ZEP_result), columns=['ToF1_Main', 'ToF2_Sub5', 'ToF3_Sub6', 'ToF4_Sub7'])
#y_ZEP_result = pd.DataFrame(y_ZEP_result, columns=['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7'])
y_ZEP_result = pd.DataFrame(y_ZEP_result, columns=['Main', 'Sub1', 'Sub2', 'Sub3', 'Sub4', 'Sub5', 'Sub6', 'Sub7'])
y_ZEP_result



#  
# (1)     
# (2)    () 
df_zero_window = df_zero_window.set_value(len(df_zero_window), 'Distance', X_count_row) #   
df_zero_window = np.array(df_zero_window)
df_zero_window_count= df_zero_window.shape[0]
df_zero_window = list(map(int, df_zero_window))



# boxflot   ( )
# 100 boxflot  
# cols_boxflot = ['min', '1st_percent', 'median', '3st_percent', 'max']
y_boxflot_col1_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col2_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col3_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col4_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col5_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col6_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col7_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col8_nonfilt = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])

# loc   +100  +99 .
for i in range (df_zero_window_count):
    y_boxflot_col1_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Main'].describe()
    y_boxflot_col2_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub1'].describe()
    y_boxflot_col3_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub2'].describe()
    y_boxflot_col4_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub3'].describe()
    y_boxflot_col5_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub4'].describe()
    y_boxflot_col6_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub5'].describe()
    y_boxflot_col7_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub6'].describe()
    y_boxflot_col8_nonfilt[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub7'].describe()

y_boxflot_col1_nonfilt = y_boxflot_col1_nonfilt.T
y_boxflot_col2_nonfilt = y_boxflot_col2_nonfilt.T
y_boxflot_col3_nonfilt = y_boxflot_col3_nonfilt.T
y_boxflot_col4_nonfilt = y_boxflot_col4_nonfilt.T
y_boxflot_col5_nonfilt = y_boxflot_col5_nonfilt.T
y_boxflot_col6_nonfilt = y_boxflot_col6_nonfilt.T
y_boxflot_col7_nonfilt = y_boxflot_col7_nonfilt.T
y_boxflot_col8_nonfilt = y_boxflot_col8_nonfilt.T



# boxflot   ( )
# 100 boxflot  
# cols_boxflot = ['min', '1st_percent', 'median', '3st_percent', 'max']
y_boxflot_col1_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col2_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col3_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col4_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col5_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col6_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col7_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
y_boxflot_col8_ZEP = pd.DataFrame(np.zeros((8,1)), index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])

# loc   +100  +99 .
for i in range (df_zero_window_count):
    y_boxflot_col1_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Main'].describe()
    y_boxflot_col2_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub1'].describe()
    y_boxflot_col3_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub2'].describe()
    y_boxflot_col4_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub3'].describe()
    y_boxflot_col5_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub4'].describe()
    y_boxflot_col6_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub5'].describe()
    y_boxflot_col7_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub6'].describe()
    y_boxflot_col8_ZEP[i] = X_Data.loc[df_zero_window[i-1]:df_zero_window[i]-1, 'Sub7'].describe()

y_boxflot_col1_ZEP = y_boxflot_col1_ZEP.T
y_boxflot_col2_ZEP = y_boxflot_col2_ZEP.T
y_boxflot_col3_ZEP = y_boxflot_col3_ZEP.T
y_boxflot_col4_ZEP = y_boxflot_col4_ZEP.T
y_boxflot_col5_ZEP = y_boxflot_col5_ZEP.T
y_boxflot_col6_ZEP = y_boxflot_col6_ZEP.T
y_boxflot_col7_ZEP = y_boxflot_col7_ZEP.T
y_boxflot_col8_ZEP = y_boxflot_col8_ZEP.T



# csv  
y_ZEP_result.to_csv("190131_RSSI data_ZEP.csv", mode='w')

y_boxflot_col1_nonfilt.to_csv("190131_RSSI data_Main_nonfilt_boxflot.csv", mode='w')
y_boxflot_col2_nonfilt.to_csv("190131_RSSI data_Sub1_nonfilt_boxflot.csv", mode='w')
y_boxflot_col3_nonfilt.to_csv("190131_RSSI data_Sub2_nonfilt_boxflot.csv", mode='w')
y_boxflot_col4_nonfilt.to_csv("190131_RSSI data_Sub3_nonfilt_boxflot.csv", mode='w')
y_boxflot_col5_nonfilt.to_csv("190131_RSSI data_Sub4_nonfilt_boxflot.csv", mode='w')
y_boxflot_col6_nonfilt.to_csv("190131_RSSI data_Sub5_nonfilt_boxflot.csv", mode='w')
y_boxflot_col7_nonfilt.to_csv("190131_RSSI data_Sub6_nonfilt_boxflot.csv", mode='w')
y_boxflot_col8_nonfilt.to_csv("190131_RSSI data_Sub7_nonfilt_boxflot.csv", mode='w')

y_boxflot_col1_ZEP.to_csv("190131_RSSI data_Main_ZEP_boxflot.csv", mode='w')
y_boxflot_col2_ZEP.to_csv("190131_RSSI data_Sub1_ZEP_boxflot.csv", mode='w')
y_boxflot_col3_ZEP.to_csv("190131_RSSI data_Sub2_ZEP_boxflot.csv", mode='w')
y_boxflot_col4_ZEP.to_csv("190131_RSSI data_Sub3_ZEP_boxflot.csv", mode='w')
y_boxflot_col5_ZEP.to_csv("190131_RSSI data_Sub4_ZEP_boxflot.csv", mode='w')
y_boxflot_col6_ZEP.to_csv("190131_RSSI data_Sub5_ZEP_boxflot.csv", mode='w')
y_boxflot_col7_ZEP.to_csv("190131_RSSI data_Sub6_ZEP_boxflot.csv", mode='w')
y_boxflot_col8_ZEP.to_csv("190131_RSSI data_Sub7_ZEP_boxflot.csv", mode='w')



#   
print('  : %.f' %X_count_row)

# Show all
#   
plt.plot(y_ZEP_result)
plt.show()

y_boxflot_col1_ZEP
```
