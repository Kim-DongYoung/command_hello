```python
import numpy as np
import matplotlib.pyplot as plt
 
START = 0
STOP = 11
INTERVAL = 0.1
 
x=[]
y=[]
for idx in range(int(START/INTERVAL),int(STOP/INTERVAL)) :
    f = np.sinc(idx*INTERVAL)
    x.append(idx*INTERVAL)
    y.append(f)
x=np.array(x)
y=np.array(y)
line0 = np.zeros_like(x)
plt.plot(x,line0,color='grey', ls=':')
plt.plot(x,line0+0.5,color='grey', ls=':')
plt.plot(x,line0+1,color='grey', ls=':')
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
START = 0
STOP = 6
INTERVAL = 0.005
SIGMA=0.1
 
cnt = 0
x=[]
y=[]
for idx in range(int(START/INTERVAL),int(STOP/INTERVAL)) :
    f = np.sinc(idx*INTERVAL)
    noise = np.random.normal(loc=0,scale=SIGMA)
    x.append(idx*INTERVAL)
    y.append(f+noise)
    cnt+=1
x=np.array(x)
y=np.array(y)
plt.plot(x,y,'b.',label="σ=%.1f N=%d"%(SIGMA,cnt))
line0 = np.zeros_like(x)
plt.plot(x,line0,color='grey', ls=':')
plt.plot(x,line0+0.5,color='grey', ls=':')
plt.plot(x,line0+1,color='grey', ls=':')
print("point num=", cnt)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
START = 0
STOP = 6
INTERVAL = 0.005
SIGMA=0.01
INCREASE = 0.00007
 
sigma = SIGMA
cnt = 0
x=[]
y=[]
for idx in range(int(START/INTERVAL),int(STOP/INTERVAL)):
    f = np.sinc(idx*INTERVAL)
    noise = np.random.normal(loc=0,scale=sigma)
    sigma += INCREASE
    x.append(idx*INTERVAL)
    y.append(f+noise)
    cnt+=1
x=np.array(x)
y=np.array(y)
plt.plot(x,y,'b.',label="σ=%.1f inc=%.5f N=%d"%(SIGMA,INCREASE, cnt))
line0 = np.zeros_like(x)
plt.plot(x,line0,color='grey', ls=':')
plt.plot(x,line0+0.5,color='grey', ls=':')
plt.plot(x,line0+1,color='grey', ls=':')
print("point num=", cnt)
plt.legend()
plt.show()
```
