```python
# OpenCV-Python Study documentation
# 이 코드는 OpenCV-Python Tutorial 을 바탕으로 작성 되었음.
# 참고 주소 : http://opencv-python.readthedocs.io/en/latest/index.html

#-*- coding:utf-8 -*-
import pandas as pd
import tensorflow as tf

import cv2
import cv2 as cv
import numpy as np
import random

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
img1 = cv2.imread('./OpenCV_data/basketball1.png',0)
img2 = cv2.imread('./OpenCV_data/basketball2.png',0)

hist1 = cv2.calcHist([img1],[0],None,[256],[0,256])
hist2 = cv2.calcHist([img2],[0],None,[256],[0,256])
```

```python
plt.subplot(221),plt.imshow(img1,'gray'),plt.title('Red Line')
plt.subplot(222),plt.imshow(img2,'gray'),plt.title('Green Line')
plt.subplot(223),plt.plot(hist1,color='r'),plt.plot(hist2,color='g')
plt.xlim([0,256])
plt.show()
```
