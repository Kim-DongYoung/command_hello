```python
# Jupyter 환경에서 코딩하는 경우, matplotlib 결과를 임베드시켜 보여주는 코드
# Jupyter 환경이 아닌 경우, 삭제하면 된다.
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
from mylib.Perceptron import perceptron

style.use('seaborn-talk')

krfont = {'family':'HYGothic-Medium', 'weight':'bold', 'size':10}
matplotlib.rc('font', **krfont)
matplotlib.rcParams['axes.unicode_minus'] = False

if __name__ == '__main__':
    style.use('seaborn-talk')
    
    df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
    
    # 아이리스 데이터를 저장한 DataFrame에서 0~99라인까지 5번째 컬럼의 데이터 값을
    # numpy 배열로 리턴
    y = df.iloc[0:100, 4].values
    
    # y에 저장된 품종을 나타내는 문자열이 'Iris-setosa'인 경우 -1, 그렇지 않은 경우
    # 1로 바꾼 numpy 배열을 다시 y에 대입.
    # 따라서 0~49번 인덱스는 -1로 50~99번 인덱스는 1로 되어 있는 numpy 배열이 됨.
    y = np.where(y=='Iris-setosa', -1, 1)
    
    # DataFrame에서 0~99라인까지 1번째, 3번째 컬럼의 데이터 값을 numpy 배열로 리턴
    # 1번째 값은 꽃받침길이이며, 3번째 값은 꽃잎길이다.
    X = df.iloc[0:100, [0, 2]].values
    
    # 필자는 '산점도'를 구하고자 하였음.
    plt.scatter(X[:50, 0], X[:50, 1], color='r', marker='o', label='setosa')
    plt.scatter(X[51:100, 0], X[51:100, 1], color='b', marker='x', label='versicolor')
    plt.xlabel('꽃잎 길이(cm)')
    plt.ylabel('꽃받침 길이(cm)')
    plt.legend(loc=4)
    plt.show()
    
    ppn = perceptron(eta=0.1)
    ppn.fit(X, y)
    print(ppn.errors_)
    
    # 머신러닝 결과
    # -0.4 + (-0.7) x (꽃받침길이) + 1.84 x (꿏잎길이) 의 값이 0보다 작거나 같으면
    # Iris-setosa, 이 값이 0보다 크면 Iris-versicolor로 분류함
    
    # 참고로 단층 퍼셉트론으 선형 분리가 가능한 데이터에 대해서만 유효함
    # 비선형적으로 분리되는 데이터에 대해서는 가중치 값을 계산하지 못하는 경우가 
    # 대부분이다.
    # 예를 들어 단층 퍼셉트론으로 AND 연산에 대한 학습은 가능하였지만
    # XOR 연산에 대한 학습은 불가능하다고 증명되었음.
    # 이러한 단층 퍼셉트론의 많은 약점을 보완하기 위해 다층 퍼셉트론 개념이 등장함.
```
